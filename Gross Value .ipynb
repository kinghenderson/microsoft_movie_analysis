```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing movie gross revenue csv
gross_rev = pd.read_csv("data/zippeddata/bom.movie_gross.csv.gz")
```

```python
# sorted by domestic gross and dropped na values 
sorted_movies_titles = gross_rev.sort_values(by='domestic_gross', ascending=False).dropna()
sorted_movies_titles
```

```python
#importing movie budgets 
movie_budget = pd.read_csv('data/zippedData/tn.movie_budgets.csv.gz')
movie_budget
```

```python
merged_df = sorted_movies_titles.merge(right= Movie_budget, how = 'inner', left_on = 'title', right_on = 'movie', suffixes=('_rounded','_budget')).head(50)
```

```python
merged_df
```

```python
# taking a column and equaling it to another variable seperate from the df
domestic_gross_rounded = merged_df.domestic_gross_rounded.head(8)
```

```python
#taking a column and equaling it to another variable seperate from the df
production_budget = merged_df.production_budget.head(8)
```

```python
#setting index equal to the movie titles from the df 
index = merged_df.title.head(8)
```

```python
graph_prod = pd.DataFrame({'domestic_gross_rounded': domestic_gross_rounded,
                   'production_budget': production_budget }, index=index)
ax = graph_prod.plot.bar(rot=100)
```

```python
#consolidating columns from merged_df to fit in bar graph easier 
data = [merged_df["domestic_gross_rounded"], merged_df["production_budget"], merged_df["title"]]
headers = ["domestic_gross_rounded", "production_budget", "title"]
df_graph = pd.concat(data, axis=1, keys=headers)
```

```python
df_graph = merged_df[["domestic_gross_rounded", "production_budget", "title"]]
```

```python
df_graph.production_budget=df_graph.production_budget.str.replace('$', '').str.replace(',','').astype(int)
```

```python
df_graph.sort_values(by='production_budget', ascending=False)
```

```python
#transfering data from graph_df into a scatter plot 
#ax = df_graph.plot.scatter(x="production_budget", y="domestic_gross_rounded")
# df_graph.plot.scatter(x='production_budget',
#                 y='domestic_gross_rounded',
#                ylim=(250, 700),
#                xlim=(30, 350)

fig, ax = plt.subplots(figsize=(14,8))
# data = df_graph[['production_budget', 'domestic_gross_rounded']]
x = df_graph['production_budget']
y = df_graph['domestic_gross_rounded']
plt.scatter(x,y)

# 
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x), "r--")

plt.title('Do Budgets influence a movies pross profit')
plt.xlabel('production_budget / 10x8 value')
plt.ylabel('domestic_gross / hundreds of millions')
plt.axvline(df_graph['production_budget'].mean(), color = 'k', label='Median')
plt.legend()
```

```python
#importing data and dropping nan values 
rotten_tomatoes = pd.read_csv('data/zippeddata/rotten_tomatoes_movies.csv.gz')
sorted_reviews = rotten_tomatoes.sort_values(by='movie_title', ascending=False).dropna()
sorted_reviews
```

```python
tomato_board = sorted_reviews[['tomatometer_rating', 'tomatometer_status', 'movie_title']]
```
